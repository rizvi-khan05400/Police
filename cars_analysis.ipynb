# Data PreProcessing

# 1. Find all Null Value in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
# Feeling the null vlaues
cars["Cylinders"].fillna(cars["Cylinders"].mean(), inplace=True)
```

```python
cars.isnull().sum()
```

# 2. Check what are the different types of "Make" are there in our dataset. And, what is the count (occurrence) of each "Make" in the data?

```python
cars["Make"].value_counts()
```

# 3. Show all the records where "Origin" is Asia or Europe.

```python
cars["Origin"].value_counts()
```

```python
cars[(cars["Origin"] == "Asia") | (cars["Origin"] == "Europe")]
```

# 4. Remove all the records (rows) where "Weight" is above 4000.

```python
cars[(cars["Weight"] <= 4000)]
```

# 5. Increase all the values of "MPG_City" column by 3.

```python
cars["MPG_City"] = cars["MPG_City"] + 3
cars
```
